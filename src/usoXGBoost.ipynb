```
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.metrics import accuracy_score
```

```
ds = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Hands on machine learning projects/PH recognition | Kaggle/pH Data.csv')
ds = ds.sample(frac=1, random_state=7).reset_index(drop=True)
```

```
from sklearn.model_selection import train_test_split
X = ds.drop('label', axis=1)
y = ds['label']
X_tr, X_ts, y_tr, y_ts = train_test_split(X,y,test_size=0.2, random_state=7)
```

```
xg_cl=xgb.XGBClassifier(n_estimators=60, max_depth=4, learning_rate=0.05, random_state=7, seed=123)
xg_cl.fit(X_tr, y_tr)
y_pred = xg_cl.predict(X_ts)
print(accuracy_score(y_ts, y_pred))
```

```
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, cross_val_score, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler

xgbCl_Pipe = Pipeline([
                      ('scaler', StandardScaler()),
                      ('XGBCl', xgb.XGBClassifier())
                     ])
params = {
          'XGBCl__n_estimators' : [25,50,75,100],
          'XGBCl__max_depth' : [4,5,6,7],
          'XGBCl__learning_rate' : [0.01,0.05,0.1,0.5,1],
          'XGBCL__colsample_bytree' : [0.3,0.7]
          }

R_xgbcl = RandomizedSearchCV(estimator=xgbCl_Pipe, param_distributions=params, n_iter=20, random_state=7, cv=10)
R_xgbcl.fit(X_tr.to_dict('records'), y_tr)

print(R_xgbcl.best_params_)   
print(R_xgbcl.score(X_tr,y_tr))                 
```
